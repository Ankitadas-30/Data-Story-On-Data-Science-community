```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
#removing the top row

df_fin=df.iloc[1:,:]
df_fin.head()
```

```python
#get percent of null values in question

#pd.set_option('display.max_rows', None)
print((df_fin.isnull().sum()/df.shape[0])*100)
```

```python
#identifying the unique question numbers present in the DataFrame's column names and storing them in a list.

#create a dictionary for questions
Questions={}
#create list of questions
qnums=list(dict.fromkeys([i.split('_')[0] for i in df_fin.columns]))# spliting on '_'
qnums
```

```python
# adding data to the respective cols
for i in qnums:
    if i in['Q1','Q2','Q3']:
        Questions[i]=df_fin[i]
    else:
        Questions[i]=df_fin[[q for q in df_fin.columns if q.startswith(i)]]
Questions['Q1']
```

# Analysis of Age(Q1)

```python
#Q1-Age

age_order = ['18-21', '22-24', '25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-69','70+']
fig=px.histogram(df_fin,x='Q1', category_orders={"Q1": age_order})
fig.update_xaxes(title_text='Age')
fig.update_layout(title='Age Distribution')
fig.show()
```

```python
#Q1- Age
#data cleaning

df_model = df_fin.dropna(subset=['Q1'])
df_model['age_cleaned'] = df_model.Q1.apply(lambda x: str(x).replace('+','').strip())
df_model.age_cleaned.value_counts()
```

```python
#Q1- Age
#Splitting the age range to min and max age

df_model['age_min'] = df_model.age_cleaned.apply(lambda x: 70 if '-' not in x else int(x.split('-')[0]))
df_model['age_max'] = df_model.age_cleaned.apply(lambda x: 70 if '-' not in x else int(x.split('-')[1]))
df_model.age_max.value_counts()
```

```python
#Q1- Age
#finding the average age

df_model['approx_age'] = (df_model.age_min+df_model.age_max)/2
df_model.approx_age.value_counts
```

```python
df_model
```

```python
#Q1 and Q5-Age by role

average_age_by_role = df_model.groupby('Q5')['approx_age'].mean().reset_index()

# Sort the DataFrame in descending order based on 'approx_Salary'
average_age_by_role = average_age_by_role.sort_values(by='approx_age', ascending=False)

fig = px.bar(average_age_by_role, x='Q5', y='approx_age', title='Average Age by Role',hover_data=['approx_age'],color_discrete_sequence=['violet'])

fig.update_traces(hovertemplate='Average Age: %{y:.0f}')
```

```python
#Q1 and Q6
#heatmap

fig=px.density_heatmap(df_fin,x='Q1',y='Q6',category_orders={'Q1':['18-21','22-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64','65-69','70+'],'Q6':['I have never written code','<1 years','1-2 years','3-5 years','5-10 years','10-20 years','20+ years']})
fig.update_xaxes(title_text='Age (Q1)')
fig.update_yaxes(title_text='Years of Coding (Q6)')
fig.show()
```

# Q7-Programming Languages usage

```python
# replacing column names with the most frequently occured values in each column
Questions['Q7'].columns=list(Questions['Q7'].mode().values)
Questions['Q7']
```

```python
#Q7
# Language used
q7=Questions['Q7'].count().reset_index()#grouping by language and aggregate function is count
q7.columns=['language','Count']
q7=q7.sort_values('Count',ascending=False)

# Calculating the total count for all languages
total_count = q7['Count'].sum()

# Calculating the percentage for each language
q7['Percentage'] = (q7['Count'] / total_count) * 100

# Creating custom hover text that includes count and percentage
hover_text = []
for index, row in q7.iterrows():
    language = row['language']
    count = row['Count']
    percentage = row['Percentage']
    hover_text.append(f"Language: {language}<br>Count: {count}<br>Percentage: {percentage:.2f}%")

# Creating the bar chart with custom hover text
fig = go.Figure(data=[go.Bar(
    x=q7['language'],
    y=q7['Count'],
    text=hover_text,
    hoverinfo='none',  # Show only the custom hover text
)])

# Customizing the layout
fig.update_layout(
    xaxis_title='Language',
    yaxis_title='Count',
    title='Language Usage Distribution',
)

fig.show()
```
